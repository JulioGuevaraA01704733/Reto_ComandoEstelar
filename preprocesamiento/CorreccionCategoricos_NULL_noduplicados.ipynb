```python
import pandas as pd

# Cargar el dataset
df = pd.read_csv('iar_Reservaciones.csv')
```

```python
df.shape
```

```python
# Verificar valores nulos 
missing_counts = df.isnull().sum()
print(missing_counts[missing_counts > 0])
```

```python
# Eliminar columnas con valores nulos
df = df.drop(columns=['h_correo_e', 'h_nom'])
```

```python
print(df.isnull().sum().sum())  # total de valores faltantes
```

```python
# Convertir valores con especios en blanco a NaN
df = df.replace(r'^\s*$', pd.NA, regex=True)
```

```python
# Nuevos valores nulos
missing_counts = df.isnull().sum()
print(missing_counts[missing_counts > 0])
```

```python
# Eliminar columnas con gran numero de valores nulos
df = df.drop(columns=['h_cod_reserva', 'h_codigop'])
```

```python
# Eliminar filas con valores nulos
df = df.dropna()
```

```python
# Verificar y eliminar filas duplicadas
duplicados = df.duplicated()
print(f"\nNúmero de filas duplicadas: {duplicados.sum()}")
df = df.drop_duplicates()
```

```python
# Nuevas dimensiones del dataset
df.shape
```

```python
for column_name in df.columns:
    value_counts = df[column_name].value_counts()
    print("Column Name: ", column_name)
    print(value_counts)
```

```python
# Convertir columnas de fecha a tipo datetime
df['h_fec_reg'] = pd.to_datetime(df['h_fec_reg'].astype(str), format='%Y%m%d')
df['h_fec_lld'] = pd.to_datetime(df['h_fec_lld'].astype(str), format='%Y%m%d')
df['h_fec_sda'] = pd.to_datetime(df['h_fec_sda'].astype(str), format='%Y%m%d')
df['h_ult_cam_fec'] = pd.to_datetime(df['h_ult_cam_fec'].astype(str), format='%Y%m%d')
df['h_res_fec'] = pd.to_datetime(df['h_res_fec'].astype(str), format='%Y%m%d')
```

```python
# Verificar el resto de valores categoricos
for column_name in df.columns:
    value_counts = df[column_name].value_counts()
    print("Column Name: ", column_name)
    print(value_counts)
```
